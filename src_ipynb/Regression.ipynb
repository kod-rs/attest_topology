```python
import pandas

data = pandas.read_csv('../datasets/kpc.csv', header=[0, 1])
data = data.dropna()
data
```

```python
import sklearn.linear_model
import sklearn.preprocessing
import sklearn.svm
import sklearn.metrics
import sklearn.pipeline
import pickle

poly = sklearn.preprocessing.PolynomialFeatures(2)

x = data.drop('goal', axis=1)
y = data.goal.p

model = sklearn.pipeline.Pipeline([
    ('poly_features', poly),
    ('lasso', sklearn.linear_model.Lasso())])
model.fit(x, y)


print(sklearn.metrics.mean_squared_error(model.predict(x), y, squared=False) / (y.max() - y.min()))

pickle.dump(model, open('../attest/estimator/p.pickle', 'wb'))
```

```python
import sklearn.linear_model
import sklearn.preprocessing
import sklearn.svm
import pickle

poly = sklearn.preprocessing.PolynomialFeatures(2)

x = data.drop('goal', axis=1)
y = data.goal.q

model = sklearn.pipeline.Pipeline([
    ('poly_features', poly),
    ('lasso', sklearn.linear_model.Lasso())])
model.fit(x, y)

print(sklearn.metrics.mean_squared_error(model.predict(x), y, squared=False) / (y.max() - y.min()))

pickle.dump(model, open('../attest/estimator/q.pickle', 'wb'))
```
