```python
import pandas

data = pandas.read_csv('../datasets/kpc.csv', header=[0, 1])
data = data.dropna()
data
```

```python
import sklearn.linear_model
import sklearn.preprocessing
import sklearn.svm

poly = sklearn.preprocessing.PolynomialFeatures(2)

x = poly.fit_transform(data.drop('goal', axis=1))
y = data.goal.p

model = sklearn.linear_model.Lasso()
model.fit(x, y)
```

```python
import sklearn.metrics


sklearn.metrics.mean_squared_error(model.predict(x), y, squared=False) / (y.max() - y.min())
```
